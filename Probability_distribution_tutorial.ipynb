```python
%matplotlib inline
import matplotlib.pyplot as plt
from IPython.display import Math, Latex
from IPython.core.display import Image
import seaborn as sns
sns.set(color_codes=True)
sns.set(rc={'figure.figsize':(5,5)})
```

# 1. 균일 분포(Uniform distribution)

```python
from scipy.stats import uniform
```

```python
n = 10000
start = 10
width = 20
data_uniform = uniform.rvs(size = n, loc = start, scale = width)
```

```python
ax = sns.distplot(data_uniform,
                 bins=100,
                 kde=True,
                 color='blue',
                 hist_kws={"linewidth":1, "alpha":0.5})
ax.set(xlabel="Uniform Distribution", ylabel="Frequency")
```

# 2. 정규 분포(Normal Distribution)

```python
from scipy.stats import norm
data_normal = norm.rvs(size=n, loc=0, scale=1)
```

```python
ax = sns.distplot(data_normal,
                 bins=100,
                 kde=True,
                 color='blue',
                 hist_kws={"linewidth":1, "alpha":0.5})
ax.set(xlabel="Normal Distribution", ylabel="Frequency")
```

# 3. 감마 분포(Gamma Distribution)

```python
from scipy.stats import gamma
data_gamma = gamma.rvs(a=5, size=n)
```

```python
ax = sns.distplot(data_gamma,
                 kde=True,
                 bins=100,
                 color='blue',
                 hist_kws={'linewidth':1, 'alpha':0.5})
ax.set(xlabel='Gamma Distribution', ylabel='Frequency')
```

# 4. 지수 분포(Exponential Distribution)

```python
from scipy.stats import expon
data_expon = expon.rvs(scale=1, loc=0, size=n)
```

```python
ax = sns.distplot(data_expon,
                 kde=True,
                 bins=100,
                 color='blue',
                 hist_kws={'linewidth':1, 'alpha':0.5})
ax.set(xlabel='Exponential Distribution', ylabel='Frequency')
```

# 5. 포아송 분포(Poisson Distribution)

```python
from scipy.stats import poisson
data_poisson = poisson.rvs(mu=3, size=n)
```

```python
ax = sns.distplot(data_poisson,
                 kde=False,
                 bins=30,
                 color='blue',
                 hist_kws={'linewidth':1, 'alpha':0.5})
ax.set(xlabel='Poisson Distribution', ylabel='Frequency')
```

# 6. 이항 분포(Binomial Distribution)

```python
from scipy.stats import binom
data_binom = binom.rvs(n=10, p=0.8, size=n)
```

```python
ax = sns.distplot(data_binom,
                 kde=False,
                 color='blue',
                 hist_kws={'linewidth':1, 'alpha':0.5})
ax.set(xlabel='Binomial Distribution', ylabel='Frequency')
```

# 7. 베르누이 분포(Bernoulli Distribution)

```python
from scipy.stats import bernoulli
data_bernoulli = bernoulli.rvs(size=n, p=0.6)
```

```python
ax = sns.distplot(data_bernoulli,
                 kde=False,
                 color='blue',
                 hist_kws={'linewidth':1, 'alpha':0.5})
ax.set(xlabel='Bernoulli Distribution', ylabel='Frequency')
```
